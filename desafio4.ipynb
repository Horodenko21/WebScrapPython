```python
from selenium import webdriver # Importa driver para navegadores pelo selenium
from selenium.webdriver.common.by import By # Importando classe By que server para selecionar os elementos
from selenium.webdriver.common.keys import Keys # Importo todas as teclas do teclado pelo pacote do Selenium
from selenium.webdriver.support.ui import Select # Import necessario para selecao de elementos em dropdown/selector
from selenium.webdriver.support.wait import WebDriverWait # Import da classe WAIT, serve para dizer o quanto se deve esperar pro navegador
from selenium.webdriver.support import expected_conditions as EC # Import de expected_conditions, e da nome reduzido (alias) EC
from dataclasses import dataclass # Import dataclass, como se fosse um objeto
import urllib.request # Import request para fazer uma requisicao para um determinado link e pegar todas as informacoes de determinado link
import os # Permite trabalhar com o Sistema Operacional
from selenium.webdriver.common.action_chains import ActionChains # Import da classe ActionChains
from pandas.io.clipboard import clipboard_set, clipboard_get # Import para mexer com menu de contexto

driver = webdriver.Chrome() # Declara que driver é do chrome
driver.implicitly_wait(time_to_wait=10)
wait = WebDriverWait(driver=driver, timeout=50, poll_frequency=0.5)
```

```python
driver.get('http://localhost:8000/#/desafio/4') # Declaro pagina que estou acessando
btnLista = driver.find_element(By.CSS_SELECTOR, '#root > div > div:nth-child(1) > button > svg')
btnLista.click()

btnPhone = driver.find_element(By.CSS_SELECTOR, r'#categoria > div.grid.grid-cols-2.xl\:grid-cols-4.gap-4 > div:nth-child(1)')
btnGame = driver.find_element(By.CSS_SELECTOR, r'#categoria > div.grid.grid-cols-2.xl\:grid-cols-4.gap-4 > div:nth-child(2)')
btnNote = driver.find_element(By.CSS_SELECTOR, r'#categoria > div.grid.grid-cols-2.xl\:grid-cols-4.gap-4 > div:nth-child(3)')
btnTVs = driver.find_element(By.CSS_SELECTOR, r'#categoria > div.grid.grid-cols-2.xl\:grid-cols-4.gap-4 > div:nth-child(4)')
```

```python
@dataclass
class phone:
    img: str
    titulo: str
    desc: str
    preco: str

@dataclass
class game:
    img: str
    titulo: str
    desc: str
    preco: str

@dataclass
class note:
    img: str
    titulo: str
    desc: str
    preco: str

@dataclass
class tv:
    img: str
    titulo: str
    desc: str
    preco: str

@dataclass
class produtos:
    celulares: list[phone]
    videoGames: list[game]
    notebooks: list[note]
    televisoes: list[tv]
```

```python
btnPhone.click()

frete = driver.find_element(By.CSS_SELECTOR, '#frete')
frete.click()

oferta = driver.find_element(By.CSS_SELECTOR, '#oferta')
oferta.click()

stars = driver.find_element(By.CSS_SELECTOR, '#four-stars')
stars.click()

rangeMin = driver.find_element(By.CSS_SELECTOR, '#price-from')
rangeMin.clear()
rangeMin.send_keys('500')

rangeMax = driver.find_element(By.CSS_SELECTOR, '#price-to')
rangeMax.clear()
rangeMax.send_keys('2500')

btnBusca = driver.find_element(By.CSS_SELECTOR, r'#filtros > div > div.inline-flex.justify-center.w-full.pb-4.mt-6.space-x-4.md\:px-4 > button.w-full.px-5.py-2.text-sm.font-medium.text-center.text-white.rounded-md.bg-gradient-to-r.from-cyan-400.to-sky-500')
btnBusca.click()

btnNextPag = driver.find_element(By.CSS_SELECTOR, r'#root > div > div.main-container > main > div.relative.bg-white.dark\:bg-gray-800.px-6.pt-10.pb-8.shadow-xl.ring-1.ring-gray-900\/5.rounded-lg.sm\:px-10 > div > div > div > div.inline-flex.w-full.gap-2.justify-between.mt-4.h-12 > button.max-w-\[250px\].bg-cyan-400.dark\:bg-slate-900.text-slate-900.dark\:text-white.py-2.px-8.rounded-sm.sm\:rounded-md.cursor-pointer.disabled\:bg-slate-100.dark\:disabled\:bg-slate-300.disabled\:cursor-not-allowed')

prods = driver.find_elements(By.CSS_SELECTOR, r'#root > div > div.main-container > main > div.relative.bg-white.dark\:bg-gray-800.px-6.pt-10.pb-8.shadow-xl.ring-1.ring-gray-900\/5.rounded-lg.sm\:px-10 > div > div > div > div.flex.flex-wrap.gap-4.max-h-\[1000px\].overflow-y-scroll.pr-8.pb-4 > div')

phones = []
phonesInPage = []

if(btnNextPag.get_attribute('disabled')):
    for prod in prods:
        img = prod.find_element(By.TAG_NAME, 'img').get_attribute('src')
        titulo = prod.find_element(By.TAG_NAME, 'h5').text
        desc = prod.find_element(By.TAG_NAME, 'p').text
        preco = prod.find_element(By.CSS_SELECTOR, 'span.my-2').text

        cel = phone(
            img=img,
            titulo=titulo,
            desc=desc,
            preco=preco
        )
        
        phones.append(cel)
else:
    while not btnNextPag.get_attribute('disabled'):

        for prod in prods:
            img = prod.find_element(By.TAG_NAME, 'img').get_attribute('src')
            titulo = prod.find_element(By.TAG_NAME, 'h5').text
            desc = prod.find_element(By.TAG_NAME, 'p').text
            preco = prod.find_element(By.CSS_SELECTOR, 'span.my-2').text

            cel = phone(
                img=img,
                titulo=titulo,
                desc=desc,
                preco=preco
            )
            
            phonesInPage.append(cel)

        phones.append(phonesInPage)

        btnNextPag.click()
```
